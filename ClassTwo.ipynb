```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Manipulations on DataFrames').getOrCreate()
```

```python
spark
```

```python
df = spark.read.option('header','true').option('inferSchema','true').csv('F1_Complete_Dataset/races.csv')
df.show(10)
```

```python
df_2020 = df.filter(df.year == 2020)
df_2020.show()
```

```python
df_2020 = df.where(df.year == 2020)
df_2020.show(10)
```

```python
df_monaco_after_2010 = df.filter( (df.year > 2010) & (df.name == "Monaco Grand Prix") )
df_monaco_after_2010.show()
```

```python
df_gp = df.where(df.name.like("%Grand Prix"))
df_gp.show()
```

```python
df_2020 = df.filter('year = 2020')
df_2020.show()
```

```python
df_2020 = df.where('year = 2020')
df_2020.show(5)
```

```python
df_test = df.where('year > 2010 and name = "Monaco Grand Prix"' )
df_test.show(10)
```

```python
df_gp = df.where('name like "%Grand Prix%" ')
df_gp.show()
```

```python
df.orderBy('year').show()
```

```python
df.orderBy(df.year.asc()).show(10)
```

```python
df.orderBy(df.year.desc()).show(10)
```

```python
df.orderBy(df.year, df.round).show(10)
```

```python
df.orderBy(df.year.desc(), df.round.asc()).show(10)
```

```python
df.orderBy("name").show(10)
```

```python
df.filter("year > 2010 and name = 'Monaco Grand Prix'").orderBy(df.raceId.desc()).show()
```

```python
df.orderBy(df.raceId.desc()).filter("year > 2010 and name = 'Monaco Grand Prix' ").show()
```

```python
from pyspark.sql.functions import col

data = [

    {"raceId": 1, "name":"Australian GP", "year":2020, "time":"06:10:00"},
    {"raceId": 2, "name": "Bahrain GP",    "year": 2021, "time": None},
    {"raceId": 3, "name": "Chinese GP",    "year": 2019, "time": None},
    {"raceId": 4, "name": "Monaco GP",     "year": 2018, "time": "14:00:00"},
]

df = spark.createDataFrame(data)
print("Original Data:")
df.show()
```

```python
print("Default Ascending on time (nulls first):")
df.orderBy("time").show()
```

```python
print("Default Ascending on time (nulls first): ")
df.orderBy(df.time.desc()).show()
```

```python
print("Ascending, but put nulls last:")
df.orderBy(df.time.asc_nulls_last()).show()
```

```python
print("Ascending but put nulls last:")
df.orderBy(df.time.desc_nulls_first()).show()
```
